```python
import numpy as np
import scipy.sparse as sp
import matplotlib.pyplot as plt
from pygsp import graphs, filters, plotting, features
```

```python
n_points = 60
dimensionality = 2
points = np.random.uniform(size=(n_points, dimensionality))
sigma=0.1
G = graphs.NNGraph(points, NNtype='knn', k=3, sigma = sigma**2, rescale=False)
H = graphs.NNGraph(points, NNtype='radius', epsilon=.12, sigma = sigma**2, rescale=False)
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
G.set_coordinates(points) #'spring'
G.plot(vertex_size=30, ax=axes[0])
H.set_coordinates(points) #'spring'
H.plot(vertex_size=30, ax=axes[1])
fig.tight_layout()
```
